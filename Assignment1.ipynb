```python
## Q3: Print the list of teams which came Top 5 in terms of points (5 points)
```

```python
df['Points'] = df.Points.astype(int)
df.sort_values(by= ['Points'], ascending= False).head().Team.tolist()
```

```python
## Q4: Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. 
##     Using the same function, find the team which has maximum and minimum goal difference. (5 points)
```

```python
def Goal_diff_count(dataframe):
    df1 = df
    df1 = df1.loc[:,['Team','GoalsFor', 'GoalsAgainst']]
    df1['GoalsDifference'] = df1.GoalsFor.astype(int) - df1.GoalsAgainst.astype(int)
    TeamwithMaxGoals = df1[df1.GoalsDifference.astype(int) == df1.GoalsDifference.astype(int).max()].Team.tolist()[0]
    TeamwithMinGoals = df1[df1.GoalsDifference.astype(int) == df1.GoalsDifference.astype(int).min()].Team.tolist()[0]
    return df1.loc[:,['Team', 'GoalsDifference']], TeamwithMaxGoals, TeamwithMinGoals

df2,TeamwithMaxGoals,TeamwithMinGoals =  Goal_diff_count(df)
print(df2)
print("Team with Max Goal Difference is " + str(TeamwithMaxGoals))
print("Team with Min Goal Difference is " + str(TeamwithMinGoals))
```
