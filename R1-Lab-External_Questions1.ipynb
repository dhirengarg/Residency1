### Find out the mean and variance for each column but for class column 

```python
Iris = Iris[['Sepal Length (in cm)',
 'Sepal Width in (cm)',
 'Petal length (in cm)',
 'Petal width (in cm)',
 'SepalExtra',
 'Class',
 ]]
```

```python
## One way of calculating 
mean = []
variance = []
for i in range(0,5):
    mean.append(Iris.iloc[:,i].mean())
    variance.append(Iris.iloc[:,i].std()*2)
print(mean)
print(variance)

```

```python
## 2nd way of calculating 
Dataset = Iris.describe()
Dataset.loc['std'] = Dataset.loc['std'].apply(lambda x: x*2)
FinalDataset = Dataset.iloc[[1,2],:]
FinalDataset.reset_index(drop=True, inplace=True)
FinalDataset.index = ['mean', 'variance']
FinalDataset
```
